## Day 26: How do I find and remove duplicate rows in pandas?

```python
import pandas as pd
```

```python
user_cols = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
# users = pd.read_table('./data/u.user', sep='|', header=None, names=user_cols, index_col='user_id')
users = pd.read_table('./data/u.user', sep='|', header=None, names=user_cols)
```

```python
users.head()
```

```python
users.set_index('user_id', inplace=True)
```

```python
users.head()
```

```python
users.shape
```

```python
# Detect duplicate zip codes
users.zip_code.duplicated()
```

```python
users[users.zip_code.duplicated()].shape
```

```python
users.zip_code.duplicated().sum()
```

```python
users.duplicated()
```

```python
users.duplicated().sum()
```

```python
users[users.duplicated()]
```

```python
users.loc[users.duplicated(),:]
```

```python
users.loc[users.duplicated(keep='first'),:]
```

```python
users.loc[users.duplicated(keep='last'),:]
# the duplicate ones are the previous ones
```

```python
users.loc[users.duplicated(keep=False),:]
```

```python
users.shape
```

```python
# drop duplicates
users.drop_duplicates(keep='first').shape
```

```python
users.drop_duplicates(keep='last').shape
```

```python
users.drop_duplicates(keep=False).shape
```

```python
# Consider only certain columns for duplicate detection
users.drop_duplicates(subset=['age', 'zip_code']).shape
```

```python
users.loc[users.duplicated(subset=['age', 'zip_code'], keep=False),:]
```
