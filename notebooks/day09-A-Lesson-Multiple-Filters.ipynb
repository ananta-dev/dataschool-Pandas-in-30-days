```python
import pandas as pd
```

```python
pd.set_option("display.max_rows", None)
```

```python
# movies = pd.read_csv('http://bit.ly/imdbratings')
movies = pd.read_csv('./data/imdb_1000.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
# movies that are at least 200 min long
movies[movies.duration>=200].head(200)
```

```python
# Long drama movies
is_drama = movies.genre == 'Drama'
is_drama.head(20)
```

```python
is_long = movies.duration>=200
is_long.head(20)
```

```python
movies[is_drama & is_long].head()
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')].head(20)
```

```python
movies[(movies.genre=='Crime') | (movies.genre=='Drama') | (movies.genre=='Action')].head(20)
```

```python
movies[(movies.genre=='Crime') | (movies.genre=='Drama') | (movies.genre=='Action')].shape
```

```python
movies.genre.isin(['Crime', 'Drama', 'Action']).shape
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].head(20)
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].shape
```
