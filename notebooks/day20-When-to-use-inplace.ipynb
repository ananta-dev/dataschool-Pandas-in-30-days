```python
import pandas as pd
```

```python
ufo = pd.read_csv('./data/ufo.csv')
```

```python
ufo.head()
```

```python
ufo.shape
```

```python
ufo.drop('City', axis=1).head()
```

```python
ufo.head()
```

```python
ufo.drop('City', axis=1, inplace=False)
```

```python
ufo.drop('City', axis=1, inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any')
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.shape
```

```python
ufo.set_index('Time', inplace=True)

# Same as 
# ufo = ufo.set_index('Time')
```

```python
ufo.head()
```

```python
# Bonus

ufo.fillna(method='bfill').tail()
```

```python
ufo.fillna(method='ffill').tail()
```

```python
ufo.iloc[:,[0,3]]
```

```python
ufo.iloc[:, 0:4]
```

```python
# iloc is exclusive of last number in range 0:3 => 0,1,2
ufo.iloc[0:3, :]
```

```python
# confusing...
ufo[['City', 'State']]
```

```python
# This is better, as it is explicit
ufo.loc[:, ['City', 'State']]
```

```python
# Confusing
ufo[0:2]
```

```python
# Better with iloc - explicit
ufo.iloc[0:2,:]
```
