```python
movies.head()
```

```python
movies.shape
```

```python
# movies that are at least 200 min long
movies[movies.duration>=200]
```

```python
is_long = movies.duration>=200
is_long
```

```python
is_long.size
```

```python
# Get the genre of the long movies
movies[movies.duration>=200].genre
```

```python
# With loc
movies.loc[movies.duration>=200, 'genre']
```
