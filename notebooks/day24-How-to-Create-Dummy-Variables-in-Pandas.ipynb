## Day 24: How do I create dummy variables in pandas?
https://courses.dataschool.io/view/courses/pandas-in-30-days/2341096-course-videos/7587845-day-24-how-do-i-create-dummy-variables-in-pandas


```python
import pandas as pd
```

```python
train = pd.read_csv('./data/titanic_train.csv')
```

```python
train.head()
```

```python
# Create dummy variable for the Sex column
train['Sex_male'] = train.Sex.map({'female': 0, 'male': 1})
```

```python
train.head()
```

```python
# A more flexible way
pd.get_dummies(train.Sex)
```

```python
pd.get_dummies(train.Sex, dtype=int)
```

```python
# Use iloc
pd.get_dummies(train.Sex).iloc[:,1:]
```

```python
pd.get_dummies(train.Sex, dtype=int).iloc[:,1:]
```

```python
# Let's add column origin information
pd.get_dummies(train.Sex, dtype=int, prefix='Sex').iloc[:,1:]
```

```python
train.Embarked.value_counts()
```

```python
pd.get_dummies(train.Embarked, dtype=int, prefix='Embarked')
```

```python
# We really only need 2 of the 3 columns:
embarked_dummies = pd.get_dummies(train.Embarked, dtype=int, prefix='Embarked').iloc[:,1:]
embarked_dummies
```

```python
# How do we attach this dataframe to the original one
train = pd.concat([train, embarked_dummies], axis=1)
```

```python
train.head()
```

```python
# How to pass a dataframe (instead of a Series) to pd.get_dummies
train = pd.read_csv('./data/titanic_train.csv')
```

```python
train.head()
```

```python
pd.get_dummies(train, columns=['Sex', 'Embarked'])
```

```python
pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True, dtype=int)
```
